# **비디오 처리**

### 비디오 읽기

```python
input_video = cv2.VideoCapture("demo.mp4")
```

```python
fourcc = int(input_video.get(cv2.CAP_PROP_FOURCC))
print("fourcc: ", fourcc)
fps = input_video.get(cv2.CAP_PROP_FPS)
print("fps: ", fps)
H = int(input_video.get(cv2.CAP_PROP_FRAME_HEIGHT))
W = int(input_video.get(cv2.CAP_PROP_FRAME_WIDTH))
print("size: ", (H, W))
```

### 비디오 보기

```python
from IPython.display import display, Image


input = cv2.VideoCapture("demo.mp4")
display_handle=display(None, display_id=True)

while input.isOpened():
  retval, frame = input.read()
  if retval:
      _, frame = cv2.imencode('.jpeg', frame)
      display_handle.update(Image(data=frame.tobytes()))
  else:
    break

input.release()
```

### 비디오 쓰기

```python
output_video = cv2.VideoWriter("asdf.mp4", cv2.VideoWriter_fourcc(*'XVID'), fps, (W, H))

while input_video.isOpened():
  retval, frame = input_video.read()
  if retval:
    output_video.write(cv2.flip(frame, 0))
  else:
    break

input_video.release()
output_video.release()
```
